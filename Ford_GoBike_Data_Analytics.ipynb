Melihat Korelasi

```python
numerical_columns = gobike.select_dtypes(include=np.number).columns
correlation_matrix = gobike[numerical_columns].corr()
sns.heatmap(correlation_matrix, annot=True)
plt.show()
```

HASIL VISUALISASI

```python
user_type_counts = gobike['user_type'].value_counts()

# Create a bar plot
plt.figure(figsize=(8, 6))
user_type_counts.plot(kind='bar')
plt.title('Jumlah Tipe User')
plt.xlabel('User Type')
plt.ylabel('Count')
plt.show()
```

```python
user_type_bike_share = gobike.groupby(['user_type', 'bike_share_for_all_trip']).size().unstack()
user_type_bike_share.plot(kind='bar', stacked=True, figsize=(8, 6))
plt.title('Korelasi Tipe User dengan Bike Share for All Trip')
plt.xlabel('User Type')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.legend(title='Bike Share for All Trip')
plt.show()
```

```python
bike_avg_duration = gobike.groupby('bike_id')['duration_sec'].mean().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
bike_avg_duration.head(10).plot(kind='bar')
plt.title('Sepeda dengan pemakaian rata-rata tertinggi')
plt.xlabel('Bike ID')
plt.ylabel('Rata-rata durasi (detik)')
plt.xticks(rotation=45)
plt.show()
```

```python
gender_counts = gobike['member_gender'].value_counts()

plt.figure(figsize=(8, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Persentase Jenis Kelamin Anggota')
plt.show()
```

```python
start_station_counts = gobike['start_station_name'].value_counts().head(10)

plt.figure(figsize=(12, 6))
start_station_counts.plot(kind='bar')
plt.title('10 Stasiun Awal Terpopuler')
plt.xlabel('Nama Stasiun Awal')
plt.ylabel('Jumlah Perjalanan')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
birth_year_counts = gobike['member_birth_year'].value_counts().head(10)

plt.figure(figsize=(12, 6))
birth_year_counts.plot(kind='bar')
plt.title('Tahun kelahiran pengguna GoBike tertinggi')
plt.xlabel('Tahun Kelahiran')
plt.ylabel('Jumlah')
plt.xticks(rotation=45, ha='right')
plt.show()
```
